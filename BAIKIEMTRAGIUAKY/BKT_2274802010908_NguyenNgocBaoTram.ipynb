```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#1. Tạo Dataframe như dữ liệu đã cho trên.
data = {
    'Name': ['Mr. Harris', 'Mrs. Bradley', 'Miss. Laina', 'Mrs. Heath', 'Mr. William', 'Mr. James', 'Mr. Timothy', 'Mrs. Oscar', 'Mrs. Nicholas', 'Miss. Marguerite', 'Mr. William', 'Miss. Hulda', 'Mr. Lawrence', 'Mr. Wilam'],
    'Height': [1.80, 1.61, 1.78, 1.74, 1.55, 1.9, 1.61, 1.40, 1.92, 1.77, 1.55, 1.85, 1.64, 1.77],
    'Weight': [90, 87, 64, 70, 100, 106, 82, 55, 77, 60, 100, 109, 80, 66],
    'AGE': [32, 45, 22, 52, 48, 21, 28, 69, 58, 21, 48, 33, 25, 70]
}
```

```python
#2. Hiển thị thông tin Dataframe vừa tạo.
df=pd.DataFrame(data)
print(df)
```

```python
#3. Tính giá trị trung bình của cột <AGE>.
df_age_mean=df['AGE'].mean()
print(df_age_mean)
```

```python
#4. Nhóm dữ liệu cột <Height> và tình tổng trung bình cột <Weight>.
df_height_for_weight=df.groupby('Height')['Weight'].mean()
print(df_height_for_weight)
```

```python
#5. Sấp xếp Dataframe theo cột <Weight> tăng dần.
df_sortweight=df.sort_values('Weight')
print(df_sortweight)
```

```python
#6. Vẽ biểu đồ cột cho <AGE>.
sns.barplot(x=df.index, y=df["AGE"])
plt.xlabel("Index")
plt.ylabel("AGE")
plt.title("Biểu đồ cột của AGE")
plt.show()
```

```python
#7. Vẽ biểu đồ phân tán cho <AGE> và <Height>.
sns.scatterplot(x=df["AGE"], y=df["Height"])
plt.xlabel("AGE")
plt.ylabel("Height")
plt.title("Biểu đồ phân tán AGE vs Height")
plt.show()
```

```python
#8. Vẽ biểu đồ đường cho cột <Weight>.
plt.plot(df["Weight"], marker="o", linestyle="-")
plt.xlabel("Index")
plt.ylabel("Weight")
plt.title("Biểu đồ đường của Weight")
plt.show()
```

```python
#9. Vẽ biểu đồ tròn cho cột <Height>..
df["Height"].value_counts().plot.pie(autopct="%1.1f%%")
plt.title("Biểu đồ tròn của Height")
plt.ylabel("")
plt.show()
```

```python
#10. Vẽ biểu đồ phân tán cho cột <Weight>,<AGE>.
sns.scatterplot(x=df["Weight"], y=df["AGE"])
plt.xlabel("Weight")
plt.ylabel("AGE")
plt.title("Biểu đồ phân tán Weight vs AGE")
plt.show()
```

```python
#11. Kiểm tra có giá trị NaN nào trong DataFrame không.
df.isnull().sum()
```

```python
#12. Tìm các giá trị trùng lặp trong DataFrame dựa trên cột 'Name' và loại bỏ các hàng trùng lặp, giữ lại hàng đầu tiên.
df.drop_duplicates(subset='Name', keep='first', inplace=True)
print(df)
```

```python
#13. Thay thế các giá trị của cột <Age> lớn hơn 40 bằng giá trị trung bình của cột đó
df.loc[df['AGE'] > 40, 'AGE'] = df_age_mean
print(df)
```

```python
#14. Tạo một cột mới trong Dataframe với tên <cs_BMI>, tính chỉ số BIM mỗi người
df["cs_BMI"] = df["Weight"] / (df["Height"] ** 2)
print(df)
```

```python
#15. Tạo cột với tên <Phan_Loai> với các mức ‘0’,’1’,’2’,’3’ tương ứng với mức chỉ số BMI trên.
def bmi_category(bmi):
    if bmi < 18.5:
        return 0  
    elif 18.5 <= bmi < 22.9:
        return 1  
    elif 23 <= bmi < 24.9:
        return 2  
    else:
        return 3  
df["Phan_Loai"] = df["cs_BMI"].apply(bmi_category)
print(df)
```

```python
#16. Chuẩn hóa giá trị của cột <cs_BMI> chuẩn hóa giá trị trong đoạn từ 0 tới 1
df["cs_BMI"] = (df["cs_BMI"] - df["cs_BMI"].min()) / (df["cs_BMI"].max() - df["cs_BMI"].min())
print(df)
```

```python
#18. Tạo cột với tên <GT>, Sau khi thực hiện, xóa các từ Mr, Mrs, Miss nếu trong cột <Name> còn các từ đó.
df["GT"] = df["Name"].apply(lambda x: "male" if "Mr." in x else "female")
df["Name"] = df["Name"].str.replace(r"Mr\.|Mrs\.|Miss\.", "", regex=True).str.strip()
print(df)
```

```python
#18. Vẽ biểu đồ tròn cho <GT>.
df["GT"].value_counts().plot.pie(autopct="%1.1f%%", colors=["pink", "blue"])
plt.title("Tỷ lệ giới tính")
plt.ylabel("")
plt.show()
```

```python
#19. Trực quan hóa dữ liệu người nữ trên <Phan_Loai>
dulieu_female = df[df["GT"] == "female"]
sns.countplot(x=dulieu_female["Phan_Loai"])
plt.xlabel("Phân loại BMI")
plt.ylabel("Số lượng")
plt.title("Phân loại BMI của nữ")
plt.show()
```
