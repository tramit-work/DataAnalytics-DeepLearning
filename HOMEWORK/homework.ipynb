```python
import pandas as pd

file_path = "/content/dirty_data.csv"
df = pd.read_csv(file_path)
df.info()
```

```python
missing_values = df.isnull().sum()
df.head(), missing_values
```

```python
df['income'] = pd.to_numeric(df['income'], errors='coerce')
df['age'].fillna(df['age'].median(), inplace=True)
df['income'].fillna(df['income'].median(), inplace=True)

df.isnull().sum
```

```python
df = df[(df['age'] >= 0) & (df['age'] <= 100)]
df = df[df['income'] >= 0]

df[['age', 'income']].describe()
```

```python
df['dob'] = pd.to_datetime(df['dob'], errors='coerce').dt.strftime('%Y-%m-%d')

df['gender'] = df['gender'].str.lower().map({'male': 'Male', 'female': 'Female', 'f': 'Female', 'fem': 'Female'})

df[['dob', 'gender']].head()
```

```python
df.drop_duplicates(inplace=True)

df.shape
```
