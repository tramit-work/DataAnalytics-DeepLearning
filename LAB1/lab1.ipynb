### CÂU 4:

```python
def thong_ke_thieu(df):
    thieu = df.isnull().sum()
    phan_tram = (thieu / len(df)) * 100
    return pd.DataFrame({'Số lượng thiếu': thieu, 'Phần trăm': phan_tram})

display(thong_ke_thieu(df))
```

###CÂU 5:

```python
df.fillna(0, inplace=True)
print("Dữ liệu sau khi xử lý:")
display(df)
```

###CÂU 6:

```python
if 'T1' in df.columns:
    df['T1'].fillna(df['T1'].mean(), inplace=True)
```

###CÂU 7:

```python
def tinh_tbm(df, cols):
    return df[cols].sum(axis=1) / len(cols)

df['TBM1'] = tinh_tbm(df, ['T1', 'L1', 'H1', 'S1', 'V1', 'X1', 'D1', 'N1'])
df['TBM2'] = tinh_tbm(df, ['T2', 'L2', 'H2', 'S2', 'V2', 'X2', 'D2', 'N2'])
df['TBM3'] = tinh_tbm(df, ['T6', 'L6', 'H6', 'S6', 'V6', 'X6', 'D6', 'N6'])
```

###CÂU 8:

```python
dieu_kien = [
    (df[['TBM1', 'TBM2', 'TBM3']].mean(axis=1) < 5.0),
    (df[['TBM1', 'TBM2', 'TBM3']].mean(axis=1) >= 5.0) & (df[['TBM1', 'TBM2', 'TBM3']].mean(axis=1) < 6.5),
    (df[['TBM1', 'TBM2', 'TBM3']].mean(axis=1) >= 6.5) & (df[['TBM1', 'TBM2', 'TBM3']].mean(axis=1) < 8.0),
    (df[['TBM1', 'TBM2', 'TBM3']].mean(axis=1) >= 8.0) & (df[['TBM1', 'TBM2', 'TBM3']].mean(axis=1) < 9.0),
    (df[['TBM1', 'TBM2', 'TBM3']].mean(axis=1) >= 9.0)
]
xep_loai = ['Y', 'TB', 'K', 'G', 'XS']
df['XL'] = pd.cut(df[['TBM1', 'TBM2', 'TBM3']].mean(axis=1), bins=[0, 5, 6.5, 8, 9, 10], labels=xep_loai)
```

###CÂU 9:

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 4))
df[['US_TBM1', 'US_TBM2', 'US_TBM3']] = scaler.fit_transform(df[['TBM1', 'TBM2', 'TBM3']])
```

###CÂU 10:

```python
df['KQXT'] = 0
df.loc[((df['DH1']*2 + df['DH2'] + df['DH3'])/4) >= 5.0, 'KQXT'] = 1
df.loc[((df['DH1'] + df['DH2']*2 + df['DH3'])/4) >= 5.0, 'KQXT'] = 1
df.loc[((df['DH1'] + df['DH2'] + df['DH3'])/3) >= 5.0, 'KQXT'] = 1
```

###CÂU 11:

```python
output_file = "/content//processed_dulieuxettuyendaihoc.csv"
df.to_csv(output_file, index=False)
print(f"Dữ liệu đã được xử lý và lưu tại {output_file}")
```
