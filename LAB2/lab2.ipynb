```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df['T1'], marker='')
plt.grid()
plt.show()
```

```python
def classify_t1(score):
    if score < 5:
        return 'k'
    elif score < 7:
        return 'tb'
    elif score < 8:
        return 'k'
    else:
        return 'g'

df['phanloait1'] = df['T1'].apply(classify_t1)
```

```python
fre_table = df['phanloait1'].value_counts()
print(fre_table)
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x=df.index, y='T1', hue='phanloait1', marker='o')
plt.ylabel('T1')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
for category in df['phanloait1'].unique():
    subset = df[df['phanloait1'] == category]
    plt.vlines(x=subset.index, ymin=0, ymax=subset['T1'], colors='r', linestyles='dotted')
plt.scatter(df.index, df['T1'], c='b', label='T1')
plt.ylabel('T1')
plt.legend()
plt.show()
```
